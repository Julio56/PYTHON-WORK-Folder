```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv("data/city_data.csv")
ride_data = pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset
city_ride_comp = pd.merge(city_data,ride_data, how="left", on=["city","city"]) 

# Display the data table for preview

city_ride_comp.head()
```

```python
#Find the Average Fare per City

Avg_fare = city_ride_comp.groupby(["city"]).mean().fare
#Avg_fare
```

```python
#Finding the City Type
#all the Urban rides
Urb_rides = city_ride_comp[city_ride_comp[("type")]=="Urban"]
#all the Suburban rides
Sub_rides = city_ride_comp[city_ride_comp[("type")]=="Suburban"]
#all the Rural rides
Rur_rides = city_ride_comp[city_ride_comp[("type")]=="Rural"]

Urb_rides

```

```python
#Finding the Total Number of Drivers by City Type


tot_driver = city_data.groupby('type').sum().driver_count
tot_driver

```

```python
#Finding the Total Number of Rides per City by Type

tot_rides = city_ride_comp.groupby('type').count().ride_id
tot_rides
```

```python
Urb_df = Urb_rides.groupby(['city']) 



Urb_avg = Urb_df ["fare"].mean()
Urb_tot_rides = Urb_df ["ride_id"].sum()
Urb_per = Urb_df ["driver_count"].min()


#Urb_per


xUrb = Urb_tot_rides


yUrb = Urb_avg

sUrb = Urb_per

#cUrb = Avg_fare[“color”]


plt.scatter(xUrb, yUrb, s = 10 * sUrb, color = "coral", alpha = 0.5, label = "Urban")
```

```python
Sub_df = Sub_rides.groupby(['city']) 



Sub_avg = Sub_df ["fare"].mean()
Sub_tot_rides = Sub_df ["ride_id"].sum()
Sub_per = Sub_df ["driver_count"].min()

xSub = Sub_tot_rides

ySub = Sub_avg

sSub = Sub_per

#cUrb = Avg_fare[“color”]


plt.scatter(xSub, ySub, s = 10 * sSub, color = "skyblue", alpha = 0.5, label = "Suburban")
```

```python
Rur_df = Rur_rides.groupby(['city']) 



Rur_avg = Rur_df ["fare"].mean()
Rur_tot_rides = Rur_df ["ride_id"].sum()
Rur_per = Rur_df ["driver_count"].min()

xRur = Rur_tot_rides

yRur = Rur_avg

sRur = Rur_per

#cUrb = Avg_fare[“color”]


plt.scatter(xRur, yRur, s = 10 * sRur, color = "gold", alpha = 0.5, label = "Rural")
```

```python
Urban = plt.scatter(xUrb, yUrb, s = 10 * sUrb, color = "coral", alpha = 0.5, edgecolors="black", label = "Urban")
Suburban = plt.scatter(xSub, ySub, s = 10 * sSub, color = "skyblue", alpha = 0.5, edgecolors="black", label = "Suburban")
Rural = plt.scatter(xRur, yRur, s = 10 * sRur, color = "gold", alpha = 0.5, edgecolors="black", label = "Rural")
# Add titles (main and on axis)
plt.xlabel("Total Number of Rides")
plt.ylabel("Average Fare")
plt.title("Pyber Ride Sharing Data")
plt.legend(handles=[Urban, Suburban, Rural], loc="best") 
plt.show()
```
